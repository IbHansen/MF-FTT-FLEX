# This notebook reads a excel sheet with data for FTT-Flex do some renaming and saves the dataframe in another excel sheet. 

```python
import pandas as pd
import re
```

```python

# Specify the file path of the Excel workbook
workbook_path = '../data/FTTAssumptionsKENYA.xlsx'

# Specify the sheet name or index of the sheet to read
sheet_name = 'ForEviews'  # You can use the sheet name or index (0 for the first sheet)


# Read the Excel sheet into a pandas DataFrame
df = (pd.read_excel(workbook_path, sheet_name=sheet_name)
      .rename(columns={'Unnamed: 0':'varname', 'Unnamed: 1':'description'})
      .dropna(subset=['varname'])
      .drop(columns='description')
      .query('not varname.str.startswith("Global") and not varname.str.startswith("Country")')
      .set_index('varname')
      .T
      .pipe(lambda dfthis: dfthis.rename(columns={c:c.replace('SIGMA_CAP__','SIGMA_CAPITAL_COST__')
                                                     .replace('SIGMA_OM__','SIGMA_OM_COST__')  
                                                     .replace('ACCUMULATED_PRODUCTION__','PRODUCTION_GLOBAL_ACCUMULATED__')  
                                                     .replace('SIGMA_FUEL__','SIGMA_FUEL_COST__')  
                                                  for c in dfthis.columns})) 
      .rename(columns= lambda x :re.sub(r'(COAL|OIL|GAS|HYDRO|SOLAR|WIND|GEOTHERMAL|NUCLEAR|IMPORTS)_PRICE',
                                                   r'PRICE__\1', x))
     )

pd.set_option('display.max_rows', None)
display(df.head().T)
```

```python
df.to_excel('digested.xlsx', index=True, na_rep='NA')
```
